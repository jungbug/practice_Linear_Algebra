```python
import numpy as np

import matplotlib.pyplot as plt

import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats("svg")
plt.rcParams.update({"font.size": 14})
```

```python
asList = [1, 2]

np.array([1, 2]) # 행 벡터

np.array([[1], [2], [3]]) # 열 벡터


```

```python
baseVectors = np.array([.75, 1])

scalars = [1, 2, 1/3, 0, -2/3]

# vector를 어떻게 그릴 것인가?
fig, axs = plt.subplots(1, len(scalars), figsize=(12, 3))

i = 0

for s in (scalars):
    v = s * baseVectors # 아마다르 곱
    axs[i].arrow(0,0, baseVectors[0], baseVectors[1],  head_width=.1, width=0.05, color="k", length_includes_head=True)
    axs[i].arrow(0,0, v[0], v[1],  head_width=.1, width=0.05, color=[0.75, 0.75, 0.75], length_includes_head=True)
    axs[i].grid(linestyle="--")
    axs[i].axis("square")
    axs[i].axis([-2, 2, -2, 2])
    axs[i].set(xticks=np.arange(-2,3), yticks=np.arange(-2,3))
    axs[i].set_title(f"$\lambda$ = {s:.2f}")
    i += 1

plt.tight_layout()
```

```python
v = np.array([1, 2])
w = np.array([4, -6])

plt.figure(figsize=(6, 6))

vPlusW = v + w

v1 = plt.arrow(0, 0, v[0], v[1], head_width=.3, width=.1, color='k', length_includes_head=True)
v2 = plt.arrow(0, 0, w[0], w[1], head_width=.3, width=.1, color=[.75, .75, .75], length_includes_head=True)
v3 = plt.arrow(0, 0, vPlusW[0], vPlusW[1], head_width=.3, width=.1, color=[.5, .5, .5], length_includes_head=True)

plt.grid(linestyle='--', linewidth=.5)
plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.legend([v1, v2, v3], ['v', 'w', 'v+w'])

plt.title("Vectors $\mathbf{v}$, $\mathbf{w}$, and $\mathbf{v}+\mathbf{w}$")
```

```python
v = np.array([1, 2])
w = np.array([4, -6])

plt.figure(figsize=(6, 6))

vPlusW = v - w

v1 = plt.arrow(0, 0, v[0], v[1], head_width=.3, width=.1, color='k', length_includes_head=True)
v2 = plt.arrow(0, 0, w[0], w[1], head_width=.3, width=.1, color=[.75, .75, .75], length_includes_head=True)
v3 = plt.arrow(w[0], w[1], vPlusW[0], vPlusW[1], head_width=.3, width=.1, color=[.5, .5, .5], length_includes_head=True)

plt.grid(linestyle='--', linewidth=.5)
plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.legend([v1, v2, v3], ['v', 'w', 'v+w'])

plt.title("Vectors $\mathbf{v}$, $\mathbf{w}$, and $\mathbf{v}-\mathbf{w}$")
```

```python
# norm 벡터의 크기
# np.random.seed(0)

v = np.random.randn(5)

normOfVec = lambda v:np.sqrt(np.sum(v ** 2))

testVec = np.array([1,2])

print(normOfVec(testVec))

np.linalg.norm(testVec)
```

```python
# 단위 벡터, 함수를 만드는데 인자로 vector 하나 받아서 그 벡터의 단위벡터를 구하는 함수

def createUnitVec(v):
    return v / np.linalg.norm(v)

v = np.array([1,2])
v2 = np.array([2,4])
v3 = np.array([3,6])

print(createUnitVec(v))
print(createUnitVec(v2))
print(createUnitVec(v3))
```

```python
# 외적 outer product
# element-wise product, Hardamard product

v = np.array([1,2,3])
w = np.array([4,5,6])

np.outer(v, w)
```

```python
# 직교벡터 분해

a = np.array([4,10])
b = np.array([5,3])

plt.figure(figsize=(6,6))

a_arrow = plt.arrow(0,0, a[0], a[1], head_width=.3, width=.1, color=[0.75,0.75,0.75], length_includes_head=True)
b_arrow = plt.arrow(0,0, b[0], b[1], head_width=.3, width=.1, color=[0.5,0.5,0.5], length_includes_head=True)

beta = np.dot(a, b) / np.dot(a,a)
vec = beta * a

vec_arrow = plt.arrow(0, 0, vec[0], vec[1], head_width=.3, width=.1, color=[0.25,0.25,0.25], length_includes_head=True)

plt.axis('square')

vecMinusB = vec - b
vecMinusB_arrow = plt.arrow(b[0], b[1], vecMinusB[0], vecMinusB[1], head_width=.3, width=.1, color="r", length_includes_head=True)

plt.text(a[0] + 0.1, a[1] , "a_arrow", fontweight="bold", fontsize=14)


plt.legend([a_arrow, b_arrow, vec_arrow, vecMinusB_arrow], ['a', 'b', 'ba', 'b - ba'])
```

```python
np.random.seed(0)

t = np.random.randn(2)
r = np.random.randn(2)

# 수평
t_para = r * (np.dot(r,t) / np.dot(r,r))

# 수직

t_perp = t - t_para

plt.figure(figsize=(7,7))

t_arrow = plt.arrow(0,0, t[0], t[1], head_width=.05, width=0.02, color=[0.5,0.5,0.5], length_includes_head=True)

r_arrow = plt.arrow(0,0, r[0], r[1], head_width=.05, width=0.02, color="orange", length_includes_head=True)

t_para_arrow = plt.arrow(0,0, t_para[0], t_para[1], head_width=.05, width=.02, color=[0.75,0.75,0.75], length_includes_head=True)
t_perp_arrow = plt.arrow(0,0, t_perp[0], t_perp[1], head_width=.05, width=.02, color=[0.75,0.75,0.75], length_includes_head=True)

plt.axis("equal")
plt.grid(linestyle="--", linewidth=.5)
plt.axis([-1, 3, -1, 3])
```
